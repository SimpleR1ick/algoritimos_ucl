```python
import random

# Função para criar o campo minado
def criar_campo_minado(dimensao, num_bombas):
    campo_minado = [[0 for _ in range(dimensao)] for _ in range(dimensao)]
    bombas_colocadas = 0
    while bombas_colocadas < num_bombas:
        x = random.randint(0, dimensao - 1)
        y = random.randint(0, dimensao - 1)
        if campo_minado[x][y] != -1:
            campo_minado[x][y] = -1
            bombas_colocadas += 1
    return campo_minado

# Função para contar o número de bombas ao redor de uma posição
def contar_bombas_vizinhas(campo_minado, x, y):
    dimensao = len(campo_minado)
    count = 0
    for i in range(max(0, x-1), min(x+2, dimensao)):
        for j in range(max(0, y-1), min(y+2, dimensao)):
            if campo_minado[i][j] == -1:
                count += 1
    return count

# Função para revelar uma posição no campo minado
def revelar_posicao(campo_minado, x, y):
    if campo_minado[x][y] == -1:
        return False
    else:
        campo_minado[x][y] = contar_bombas_vizinhas(campo_minado, x, y)
        return True

# Função para imprimir o campo minado
def imprimir_campo_minado(campo_minado):
    dimensao = len(campo_minado)
    for i in range(dimensao):
        for j in range(dimensao):
            if campo_minado[i][j] == -1:
                print("*", end=" ")
            else:
                print(campo_minado[i][j], end=" ")
        print()

# Criar campo minado
dimensao = 5
num_bombas = 5
campo_minado = criar_campo_minado(dimensao, num_bombas)

# Jogar o campo minado
jogando = True
while jogando:
    imprimir_campo_minado(campo_minado)
    x = int(input("Digite a linha: "))
    y = int(input("Digite a coluna: "))
    jogando = revelar_posicao(campo_minado, x, y)
    if not jogando:
        print("BOOM! Você perdeu!")
        break

    # Verificar se todas as posições livres foram abertas
    count = sum(row.count(0) for row in campo_minado)
    if count == num_bombas:
        print("Parabéns! Você venceu!")
        jogando = False
```
